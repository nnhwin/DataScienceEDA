```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('dataset/Sales_Data/Sales_January_2019.csv')
df
```

```python
df.isnull().any()
```

```python
df.dropna(inplace=True)
```

```python
df.dtypes
```

```python
df=df[df['Order Date']!='Order Date']
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Price Each']=pd.to_numeric(df['Price Each'])
df['Quantity Ordered']=pd.to_numeric(df['Quantity Ordered'])
```

```python
df.dtypes
```

```python
df['Month']=df['Order Date'].dt.month_name()
```

```python
df.head()
```

```python
df['Sales Amount']=df['Quantity Ordered']*df['Price Each']
```

```python
df.head()
```

```python
#univariate analysis
sns.catplot(x='Month',data=df,kind='count')
```

```python
sns.kdeplot(x='Sales Amount',data=df)
```

```python
df.columns
```

```python
df['City']=df['Purchase Address'].apply(lambda x:x.split(',')[1].strip())
```

```python
df.head()
```

```python
sns.catplot(data=df,x='City',y='Sales Amount',palette='bright',col='Product',kind='bar',col_wrap=3)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.heatmap(df.corr(),annot=True,cmap="YlGnBu")
```

```python
df.groupby('Month')['Sales Amount'].mean().plot(kind='bar')
```

```python
df.City.value_counts()
```

```python
ind=df.Product.value_counts().index
plt.figure(figsize=(20,8))
boston=df[df.City=='Boston'].groupby("Product")['Sales Amount'].mean()
sns.lineplot(x=ind,y=boston.values,label='boston',color='red')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```
